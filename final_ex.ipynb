```python
import pandas as pd
```

```python
df=pd.read_csv('~/Desktop/Flatiron/genres2.csv') # got a new file csv to easy to look
```

```python
df.head()
```

```python
newlist=[]
for x in df.loc[:, 'genre_ids']:
    a=x[1:-1].split(', ')
    if not a in newlist:
        newlist.extend(a) #make a list with all the genres
```

```python
new_set=set(newlist)#make a set for unique genre in set
```

```python
for y in new_set:#create the columns with each genre and check on each movie
    df[y]=df['genre_ids'].map(lambda x : y in x)
```
