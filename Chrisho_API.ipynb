```python
import pandas as pd
import json
import requests
```

```python
def get_keys(path):
    with open(path) as f:
        return json.load(f)
```

```python
key=get_keys('/Users/hochinh/.secret/themovie_api.json')
api_key=key['api_key']
```

```python
res1=requests.get('http://api.themoviedb.org/3/discover/movie?&sort_by=popularity.desc&page=1&api_key='+api_key).json()
```

```python
newlist=[]
n=0
while n<50:
    n+=1
    url='http://api.themoviedb.org/3/discover/movie?&sort_by=popularity.desc&offset=20&page={}&api_key='.format(n)
    req=requests.get(url+api_key).json()
    results=req['results']
    newlist.extend(results)
```
